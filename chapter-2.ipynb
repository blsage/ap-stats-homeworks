```r
temps <- read.csv("./data-tables/ex02_085.csv")
xyplot(Temp ~ Month, data=temps)
```

There seems to be a strong association between the variables, but it certainly does not seem to be linear.

### b.

```r
model <- lm(Temp ~ Month, data=temps)
coe <- coef(model)
ladd(panel.abline(a=coe[1], b=coe[2], col="green"))
```

```r
temp.fit <- makeFun(model)
temps$residuals <- temps$Temp - temp.fit(temps$Month)
xyplot(residuals ~ Temp, data=temps)
```

Since the residual plot is somewhat linear, that means that the data are not best fit to a linear model. If the data are best fit to this linear model, then the residual plot should look random.
